Report files manager



```python
import os
import pandas as pd

already_checked = []

existing_file = False

def scroll_reports():

  folder_path = "/teamspace/studios/this_studio/.local/share/garak/garak_runs/"

  for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    #print(file_path)
    if os.path.isfile(file_path):
      if filename not in already_checked and filename.endswith("report.jsonl"):
        #print(filename)
        already_checked.append(filename)
        df = read_json(file_path)
        #print(df["Probe"])
        output_path = './extracted.xlsx'

        startrow = 1

        if not os.path.exists(output_path):
            # Create an empty Excel file if it doesn't exist
            pd.DataFrame().to_excel(output_path, sheet_name='Foglio1', index=False)
            startrow = 1
        else:
            existing_file = True
            existing_data = pd.read_excel(output_path, sheet_name='Foglio1', engine='openpyxl')
            if existing_data.empty or existing_data.isnull().all().all():
                startrow = 1
            else:
                startrow = len(existing_data) + 1

        if existing_file:
          df_excel = pd.read_excel(output_path, engine='openpyxl')

          index_to_drop = set()

          cmp_first_n = 5

          for r_excel in df_excel.itertuples(index=False, name=None):
              for pos, r_df in enumerate(df.itertuples(index=False, name=None)):
                  if r_excel[:cmp_first_n] == r_df[:cmp_first_n]:
                      index_to_drop.add(df.index[pos])
                      break

          if index_to_drop:
              df = df.drop(index=index_to_drop)



        with pd.ExcelWriter(output_path, engine='openpyxl', mode='a', if_sheet_exists='overlay') as writer:
          df.to_excel(writer, sheet_name='Foglio1', index=False, header=False if startrow > 1 else True, startrow=startrow)


          worksheet = writer.sheets['Foglio1']
          for column in worksheet.columns:
              max_length = 0
              column_letter = column[0].column_letter
              for cell in column:
                  try:
                      if len(str(cell.value)) > max_length:
                          max_length = len(str(cell.value))
                  except:
                      pass
              adjusted_width = (max_length + 2)
              worksheet.column_dimensions[column_letter].width = adjusted_width
```
